```python
import os 
import argparse
from plantcv import plantcv as pcv

class options:
    def __init__(self):
        self.image = "/Users/garethtate/plantcv/Still-1-Camera_Name-2021-06-13_11-27-57.jpg"
        self.debug = "plot"
        self.writeimg= False 
        self.result = "vis_nir_test_results.json"
        self.outdir = "/Users/garethtate/plantcv/naives_bayes_images/" # Store the output to the current directory
        self.pdfs = "/Users/garethtate/plantcv/naive_bayes_pdfs.txt"

# Get options
args = options()

# Set debug to the global parameter 
pcv.params.debug = args.debug
```

```python
# Read in the input image
vis, path, filename = pcv.readimage(filename=args.image)

# Classify each pixel as plant or background (background and system components)
masks = pcv.naive_bayes_classifier(rgb_img=vis, pdf_file=args.pdfs)
colored_img = pcv.visualize.colorize_masks(masks=[masks['Plant'], masks['Background']], 
                                               colors=['green', 'blue'])
```
